```python
import pandas as pd 
import numpy as np 
```

```python
class Perceptron():

    def __init__(self):
        self.weights = []

    def fit(self, X, y, learning_rate=0.01, num_iteration=100):
        (n_row , n_feat) = X.shape

        # initializing weights 
        self.weights = np.random.rand(n_feat+1)
        
        for i in range(num_iteration):
            
            # 1. selecting a row at random
            r_i = np.random.randint(0,n_row-1)
            row = X[r_i,:]

            # 2. get prediction for that row
            yhat = self.predict(row)

            # 3. calculate error
            error = (y[r_i] - yhat)

            # 4. set_bias
            self.weights[0] = self.weights[0] + learning_rate*error

            # 5. update weights for each feature
            for f_i in range(n_feat):
                self.weights[f_i] = self.weights[f_i] + learning_rate*error*row[f_i]

            # print weights after each iteration
            if i % num_iteration == 0:
                total_error = 0 
                for r_i in range(n_row):
                    yhat = self.predict(X[r_i,:])
                    total_error += (y[r_i] - yhat)**2

                mean_error = total_error/n_row
                print(f"Iteration {i} , Mean Squared Error: {mean_error}")

    def predict(self, row):

        # 1. start with bias
        activation = self.weights[0]

        # 2. add weighted sum of features
        for f_i in range(len(row)):
            activation += self.weights[f_i+1] * row[f_i]
        return 1 if activation >= 0 else 0
    

```
